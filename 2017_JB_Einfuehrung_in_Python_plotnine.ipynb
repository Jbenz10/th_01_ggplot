# plotnine

Jedoch haben Nutzer, die von R zu Python gewechselt sind häufig die Möglichkeiten von ggplot2 vermisst. Nun, dank plotnine, können die Nutzer gewohnte Anwendungen aus ggplot2 auch auf der Programmieroberfläche Pyhton anwenden, indem sie die Bibliothek "plotnine" nutzen.  

Es können wie in ggplot2 Grafiken und Diagramme erzeugt werden, ohne großen Aufwand. 

Dieses Notebook stellt eine Art Vergleich zu ggplot2 dar, da mit den selben Datensätzen die gleichen Grafiken erstellt werden. Es dient dazu, dass die Nutzer ein Gefühl dafür bekommen, wie ähnlich diese Programme sind. 

Die verschiedenen Unterschiede zwischen ggplot2 und plotnine werden direkt an den Anwendungsbeispielen genannt.

## Anwendungsbeispiele 

Plotnine kann am besten erlernt werden durch "learning by doing". Deshalb werden verschiedene Methoden und Befehle anhand von Anwendungsbeispielen generiert.

**Beachte:** 
Beim Importieren erscheint eine "FuturWarning" diese kann ignoriert werden. 

```python
# Wichtige Pakete und Anwendungen werden importiert 

import pandas as pd
import numpy as np

from plotnine import *
```

Wie bei ggplot2 wird nun der gleiche Datensatz eingelesen. 

```python
from_csv = pd.read_csv('data/mariano-rivera.csv',  sep=';')
# Zeige mir die ersten 5 Zeilen des Datensatzes an
from_csv.head()
```

### Streudiagramm 

Wie in ggplot2 wird als Erstes ein Streudiagramm erzeugt. Dies gilt als einer der leichtesten Diagramme, sowohl beim Erstellen als auch beim Interpretieren. 
Auf die Interpretation wird sowohl in ggplot2 als auch bei plotnine komplett verzichtet, da das Coding im Vordergrund steht. 

#### Unterschied zu ggplot2: 
1. Der komplette Befehl muss umklammert sein, d.h. *(Code)*
2. Die Variablen (Spaltennamen) müssen bei plotnine in **''** stehen, bei ggplot2 muss der X-Achse der Wert zugewiesen werden ( x= ..., y=...). Plotnine weiß, dass der erste Wert für die X-Achse bestimmt ist und der zweite Wert für die Y-Achse. 

#### Gemeinsamkeiten 
1. Beide Bibliotheken verwenden den Befehl *ggplot* und *geom_point()*, dies liegt daran, dass plotnine auf ggplot2 aufbaut, daher wurden die Befehle übernommen, um dem Nutzer eine Einfachheit zu gewährleisten 
2. In beiden Fällen werden die Farben mit *color='...'* erzeugt. 

**Anmerkung:** In beiden Fällen ist es egal, ob *colour()* oder *color()* geschrieben wird, sowohl amerikanisches Englisch als auch britisches Englisch kann verwendet werden. 

Es soll an dieser Stelle ein Streudiagramm erzeugt werden mit den Spalten "Age" und "BB". 
Schnell ist zu sehen, dass die gleichen Befehle dafür benutzt werden wie in R, Unterschiede wurden bereits oben erläutert. 

```python
(ggplot(from_csv, aes('Age', 'BB')) + 
 geom_point(color='red'))
```

### Regressionsgerade

#### Unterschiede zu ggplot2: 
1. Wie bereits beim Streudiagramm erwähnt, müssen die Werde in **'...'** stehen (Dies ist bei jedem Diagramm der Fall, deshalb wird in den folgenden Diagrammen dieser Unterschied nicht nochmals erwähnt) 
2. Bei der Zuweisung der Daten(from_csv) wird bei ggplot2 folgendes geschrieben: "data=from_csv", bei Plotnine ist dies nicht notwendig, es wird nur der Dateiname geschrieben 

#### Gemeinsamkeiten 
Die Befehle sind an dieser Stelle gleich

Es wird eine Regressionsgerade mit eingebaut, der dafür verwendete Befehl ist 'lm'. Es wird zusätzlich die Glättung durchgeführt mit Berücksichtigung des 95%-Intervalls. 

```python
pplot = ggplot(from_csv, aes('Age', 'BB')) + geom_point()
# geom_smooth (Glättung) wird hinzugefügt, mit "lm" wird die Regressionsvariable erzeugt, Intervall ist mit enhalten
pplot + geom_smooth(method = 'lm', se='F')
```

### Liniendiagramm 

Hier wurde nicht der gleiche Datensatz verwendet wie in ggplot2. 

Grund: 
1. Der Befehl "np.random.randint" dieser benutzt die Bibliothek numpy (oben eingelesen) und generiert bei jedem durchlauf eine Zufällige Zahlenfolge und schreibt diese in den Dataframe rein. Dabei sollen die Zahlen zwischen 0 und 10 sein. 

```python
df2 = pd.DataFrame(np.random.randint(low=0, high=10, size=(5, 5)),
                    columns=['a', 'b', 'c', 'd', 'e'])
df2


# Ein normaler ggplot mit Basis Informationen werden erstellt
pplot = ggplot(df2, aes('a', 'b'))

# geom_line ( die Linien ) werden mit dem + Operator angehängt, Ergebnis -> Liniengrafik 
pplot +  geom_line() 

```

```python
pplot2= ggplot(df2, aes('c', 'd'))

# Es werden diesesmal die Zeilen c und d für die Grafik verwendet 
# Und die Farbe auf "Orange" gesetzt
pplot2 + geom_line(color="orange")
```

### Balkendiagramm

### Heatmap 
Daher wird nur noch eine Heatmap erzeugt. Dafür wird der Datensatz "flights.csv" verwendet. 

```python
# Einlesen der Datei
flight = pd.read_csv('data/flights.csv')
months = flight['month'].unique()  # Monate werden sortiert und nur die ersten 5 verschiedenen 
# unique = Redundanz (keine gleichen Datzensätze)
flight['month'] = pd.Categorical(flight['month'], categories=months)
flight.head()
```

#### Unterschied zu ggplot2 
Abgesehen von der Eingabe der Werte und die äußerste Klammer gibt es nur einen wichtigen Unterschied, die Farben 
- Bei ggplot2 ist die Standardfarbe immer Blau, ein Farbverlauf von verschiedenen Blautönen (von dunkel zu hell) 
- Bei plotnine ist die Standardfarbe ein Farbverlauf von Blau ins grünliche bis hin zu gelb. Um die Unterschiede noch mehr zu verdeutlichen. 

```python
# factor(year) = Dient dazu, dass alle Jahre eingezeigt werden und nicht nur ein paar wenige, an erster Stelle -> X-Achse
# Die .95s erzeugen die Lücken zwischen den Kacheln 
# geom_text = erzeugt den Text der Kacheln, wird gefüllt mit den Daten aus der Spalte "Passengers"
# size = die Größe der Zahlen in den Kacheln 
# *Anmerkung*: Die Farbei des Ergbnisses ist in plotnine so als Standard 

(ggplot(flight, aes('factor(year)', 'month', fill='passengers'))
 + geom_tile(aes(width=.95, height=.95))
 + geom_text(aes(label='passengers'), size=10)
)
```

```python
(ggplot(flight, aes('factor(year)', 'month', fill='passengers'))
 + geom_tile(aes(width=.95, height=.95))
 + geom_text(aes(label='passengers'), size=10, color='white')  # Die Farbe der Zahlen wird auf weiß gesetzt
)
```

```python
# text_c = Die Farbe des Textes, das Array bekommt die Farbe Schwarz in der Länge des Datensatzes 
# Die Farbe wird auf Weiß gesetzt, sobald der Wert der Variable <300, ansonsten ist die Farbe Schwarz

text_c = np.array(['black']*len(flight))
text_c[flight['passengers']<300] = 'white'
```

```python
# Es wird eine Art Heatmap gezeichnet, genau so wie oben 
# Bei color=text_c wird die oeben definierte Farbe eingefügt. 

(ggplot(flight, aes('factor(year)', 'month', fill='passengers'))
 + geom_tile(aes(width=.95, height=.95))
 + geom_text(aes(label='passengers'), size=10, color=text_c)  # modified
)
```
